we can handle error by creating exception by 3 key worlds
- try
- catch
- finally

```python
#basic format
x = True
try:
    #syntax error its will not exeute.
    #print("ddfdf)
    lis = [1,2,3]
    print(lis1)
    x=2/0
    for i in range(0,len(lis)):
        print(lis[i+10])
except IndexError:
    print("\n the element is not present in the list")
    x=False
    print(x)
except ZeroDivisionError:
    print("\n any number divided by zero is infinity")
except NameError:
    print("\n No such variable present")
    x=False
    print(x)
except:
    print("\n wrong syntex used")
finally:
    print("rest of the statements")
```

# Regular Expressions

If you would like to find any pattern in the text that time we use regular expressions

```python
text = "The phone number of agent is is 444-555-6789 ,please call by the phone only?"
```

```python
"444-555-6789" in text
```

### If there is a match

Patter you want to search

```python
import re
patter = "phone"
```

note :- The re.search method ,it extract the first match of in the string

```python
#it return a re object
match_string=re.search(patter,text)
match_string
```

note :- To find all the match 's in the string ,we can use find all method ,where this method returns list of matched string

```python
match_string = re.findall(patter,text)
match_string
```

Note :- If you want to get all the matches as an re object

```python
for i in re.finditer(patter,text):
    print(i.span())
```

#### searching by patterns

previously we have search by the basic string now,we use patterns to search some patterns in the strings

#### notations

#### identifiers 

```python
from IPython.display import Image
Image("pic1.png")
```

```python
#trying to extract a phone number from a string
text = "This is my phone number 100-100-1001 ,and there is one more alternative number 999-999-9999"
patter = r"\d\d\d-\d\d\d-\d\d\d\d"
```

```python
#for getting the string
re.search(patter,text).group()
```

```python
#for getting the mathed strings
re.findall(patter,text)
```

```python
#for getting all the match object
for i in re.finditer(patter,text):
    print(i.group())
```

##### Other special operations

extracting each part of the match by keeping sepration in the expression

```python
from IPython.display import Image
Image("pic3.png")
```

```python
text = "This is my phone number 100-100-1001 ,and there is one more alternative number 999-999-9999"
patter = r"(\d{3})-(\d{3})-(\d{4})"
#extracting each piece of match by indexing
print(re.search(patter,text).group(1))
print(re.search(patter,text).group(2))
print(re.search(patter,text).group(3))
```

```python
pattern = r"[^+.""'',?]+"
sentence = "My name is Rohan.I am a DataScientist and I love Analytics and AI .What do you like?"
" ".join(re.findall(pattern,sentence))
```
